# -------------------------------------- 4. Transformation ----------------------------------------------


üìå4.1 Concater les deux df animaux et v√©g√©taux

Ajouter un columns suppl√©mentaire pour l'identification

```python
anim_clean['Origine'] = 'animaux'
veg_clean['Origine'] = 'vegetaux'
```

```python
anim_veg = pd.concat([anim_clean, veg_clean], ignore_index = True) # pour ignorer les valeurs de l'index dans la concat√©nation
anim_veg.head()
```

```python
anim_veg['Unit√©'].unique()
```

```python
anim_veg.T
```

üìå4.2 Pivoter le df pour avoir une forme manipulable et compatible avec notre besoin

```python
anim_veg = anim_veg.pivot_table(index = ['Zone','Code_zone','Produit', 'Origine'], values='quantite',columns=['√âl√©ment'])
anim_veg = anim_veg.rename_axis(None, axis = 1)
anim_veg.reset_index(inplace=True)

#Remplacer les 0 par nan pour √©viter l'erreur
anim_veg.replace( 0.0 , np.nan, inplace = True) 

# Changement les noms de columns
anim_veg.rename(columns = {'Disponibilit√© alimentaire (Kcal/personne/jour)':'Disponibilit√©_alimentaire(Kcal/personne)', 'Disponibilit√© de mati√®re grasse en quantit√© (g/personne/jour)':'Disponibilit√©_de_mati√®re_grasse_en_quantit√©(kg/personne)', 'Disponibilit√© de prot√©ines en quantit√© (g/personne/jour)':'Disponibilit√©_de_prot√©ines_en_quantit√©(kg/personne)', 'Disponibilit√© alimentaire en quantit√© (kg/personne/an)':'Disponibilit√©_alimentaire_en_quantit√©(kg/personne/an)'}, inplace = True)

anim_veg.T
```

```python
anim_veg.info()
```

üìå4.3 merger les dfs anim_veg avec population

```python
pop_zone_produit = pd.merge(anim_veg, pop_clean[['Zone','population']], how = 'inner', on = ['Zone'])

#Remplacer les 0 par nan pour √©viter l'erreur
pop_zone_produit.replace( 0.0 , np.nan, inplace = True) 

pop_zone_produit
```

```python
plt.figure(figsize = (10, 5))
sns.heatmap(pop_zone_produit.isna(), cbar = False)
```
